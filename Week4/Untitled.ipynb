```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gofr = np.loadtxt('es4.3/gas/output.gofr.0')
nsteps = len(gofr[0])
nbins = len(gofr)
g_ave = np.zeros(nbins)
for ibin in range(nbins):
    for istep in range(nsteps):
        g_ave[ibin]+= gofr[ibin][istep]
    g_ave[ibin]/=nsteps
g_ave = np.asarray(g_ave)
plt.plot(np.arange(nbins),g_ave)
```
